#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
# Instantiate variables
drawing = True
ix = -1
iy = -1

def draw_rectangle(event, x, y, flags, params):
    global drawing, ix, iy, mode

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y

    elif event == cv2.EVENT_MOUSEMOVE:
       if drawing == True:
            cv2.rectangle(img, (ix, iy), (x, y), (255, 0, 0), 3)

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img, (ix, iy), (x, y), (255, 0, 0), 3)

image = cv2.imread('../DATA/dog_backpack.jpg')      
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.namedWindow('man and dog')
cv2.setMouseCallback('man and dog', draw_rectangle)

while True:
    cv2.imshow('man and dog', img)

    if cv2.waitKey(0) & 0xFF == 27:
        break

cv2.destroyAllWindows()
     
        
    


```
